## Tokenizing

```python
%pip install transformers
```

```python
from transformers import AutoTokenizer
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    pretrained_model_name_or_path='google/bert_uncased_L-2_H-128_A-2')
```

```python
train_dataset = cola_dataset['train']
val_dataset = cola_dataset['validation']
test_dataset = cola_dataset['test']
```

```python
tokenizer
```

```python
print(train_dataset[0]['sentence'])
tokenizer(train_dataset[0]['sentence'])
```

```python
tokenizer.decode(tokenizer(train_dataset[0]['sentence'])['input_ids'])
```

```python
def encode(example):
    return tokenizer(example['sentence'],
                     truncation=True,
                     padding='max_length',
                     max_length=512)
```

```python
train_dataset = train_dataset.map(encode, batched=True)
train_dataset
```

```python
train_dataset[0]
```
